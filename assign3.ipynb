```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Scatter Plot")
plt.show()
```

```python
fig,axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
axes[0, 0].plot(x1, y1, color='blue')
axes[0, 0].set_title('1')

axes[0, 1].plot(x2, y2, color='green')
axes[0, 1].set_title('2')

axes[1, 0].plot(x3, y3, color='red')
axes[1, 0].set_title('3')

axes[1, 1].plot(x4, y4, color='purple')
axes[1, 1].set_title('4)')

plt.show()
```
